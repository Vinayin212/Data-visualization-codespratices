### Matplotlib basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
x
y= x+20
y
```

```python
plt.plot(x, y)
plt.title("income vs expenses")
plt.xlabel("income")
plt.ylabel("expenses")    
```

### subplots in matplotlib

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title("income vs expenses")
plt.xlabel("income")
plt.ylabel("expenses")
plt.subplot(2, 2, 2)
plt.plot(y, x)
plt.title("expenses vs income")
plt.xlabel("expenses")
plt.ylabel("income")
```

### Object Oriented way

```python
fig = plt.figure(figsize=(8, 8), dpi=100)

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x, y)
axis2 = fig.add_axes([0.2, 0.6, 0.2, 0.2])
axis2.plot(y,x)
```

### types of plots

```python
plt.scatter(x, y)
plt.title("income vs expenses")
plt.xlabel("income")
plt.ylabel("expenses")
```

```python
plt.hist(x)
```

```python
plt.boxplot(x)
```

### saving plots

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, y)
plt.savefig("my_plot.png")
```

### working with images

```python
import matplotlib.image as mpimg
img = mpimg.imread("my_plot.png")
plt.imshow(img)
plt.axis("off")
```

### designing

```python
plt.plot(x, y, color="red", linestyle="--", marker="o", markersize=10, markerfacecolor="blue", markeredgewidth=2, markeredgecolor="green")
plt.title("income vs expenses")
plt.xlabel("income")    
plt.ylabel("expenses")
```

### matrixplots

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

#### heatmaps

```python
tipscorr = tips[['total_bill', 'tip','size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

#### clustermap

```python
sns.clustermap(tipscorr.corr())
```

#### pivot tables heat map

```python
pvtflight = flights.pivot_table(values= 'passengers', index='month',columns='year')
pvtflight
```

```python
sns.heatmap(pvtflight)
```
